```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
X = df.iloc[:, [3,4]].values
```

```python
# Using the elbow method to find optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
# Now to fit the model to the data with the optimal no. of cluster
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means = kmeans.fit_predict(X)
```

```python
y_means
```

```python
# Visulaizing the clusters formed
plt.scatter(X[y_means==0, 0], X[y_means==0, 1], s=99, color='magenta', label='Cluster 1')
plt.scatter(X[y_means==1, 0], X[y_means==1, 1], s=99, color='red', label='Cluster 2')
plt.scatter(X[y_means==2, 0], X[y_means==2, 1], s=99, color='yellow', label='Cluster 3')
plt.scatter(X[y_means==3, 0], X[y_means==3, 1], s=99, color='cyan', label='Cluster 4')
plt.scatter(X[y_means==4, 0], X[y_means==4, 1], s=99, color='green', label='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=299, color='orange', label='Centroids')
plt.title('Clusters of Clients')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()s
plt.show()
```
